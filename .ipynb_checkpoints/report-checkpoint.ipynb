```python
import time
```

```python
def tail_f(file):
    interval = 1.0
    while True:
        where = file.tell()
        line = file.readline()
        if not line:
            time.sleep(interval)
            file.seek(where)
        else:
            yield line
```

```python
for line in tail_f(open('data/test_runs.log')):
    print(line)
```
